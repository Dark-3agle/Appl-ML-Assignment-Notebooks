# Preparing and Calculating all Required rows for the Second Part

## Calculating Mode

```python
# 
n_mode=pd.DataFrame(df.mode(axis=0))
n_mode=n_mode[:1]
n_mode_t=n_mode.transpose()
n_mode_t.columns=["MODE"]
n_mode_t
n_mode_new=n_mode_t.transpose()
n_mode_new
Part_1_df=Part_1_df.append(n_mode_new)
Part_1_df
```

## Calculating number of Zeros

```python
N_Zero=pd.DataFrame((df==0).sum(axis=0))
N_Zero.columns=["NZERO"]
N_Zero_t=N_Zero.transpose()
Part_1_df=Part_1_df.append(N_Zero_t)
Part_1_df
```

## Calculating Kurtosis

```python
Kurt=pd.DataFrame(df.kurtosis(axis=0))
Kurt.columns=["KURT"]
Kurt
kurt_t=Kurt.transpose()
kurt_t
Part_1_df=Part_1_df.append(kurt_t)
Part_1_df
```

## Calculating Cardinality

```python
Cardinality=pd.DataFrame(df.nunique())
Cardinality.columns=["CARDINALITY"]
Cardinality=Cardinality.transpose()
Cardinality
Part_1_df=Part_1_df.append(Cardinality)
Part_1_df
```

## calculating Number of missing values

```python
n_missing=pd.DataFrame(df.isnull().sum(axis=0))
n_missing.columns=["NMISSING"]
n_missing
n_missing_t=n_missing.transpose()
n_missing_t
Part_1_df=Part_1_df.append(n_missing_t)
Part_1_df
```
